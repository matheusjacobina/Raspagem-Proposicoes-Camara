```python
# importação de bibliotecas

import pandas as pd
import requests
from datetime import date
from datetime import timedelta

# palavras-chaves previamente definidas e já no formato devido.
# Para adicionar novas, usamos o site dados abertos da Câmara para gerar o link
# e assim copiar a parte com os 'keywords'.

# No momento, utilizamos as seguintes palavras-chaves:
# "Comissão de Valores Mobiliários" ; "6.404" ; "6.385" ; "Bolsa de Valores"
# "Mercado de Capitais" ; "Fundos de Investimentos" ; "Agentes Autônomos"
# "CVM" ; "Dividendos" ; "Debêntures" ; "Valor Mobiliário"

keywords = """\
keywords=comiss%C3%A3o%20de%20valores%20mobili%C3%A1rios&\
keywords=6.404&\
keywords=6.385&\
keywords=bolsa%20de%20valores&\
keywords=mercado%20de%20capitais&\
keywords=mercado%20de%20a%C3%A7%C3%B5es&\
keywords=fundos%20de%20investimentos&\
keywords=agentes%20aut%C3%B4nomos&\
keywords=cvm&\
keywords=dividendos&\
keywords=deb%C3%AAntures&\
keywords=valor%20mobili%C3%A1rio&"""

# A partir daqui, inserimos o intervalo das datas de apresentação procuradas.
# A ideia e ter uma parte automática, onde o programa pegue a data de "hoje" e
# busque por proposições em um período de 5 dias para trás.

# A outra parte se dá a opção do usuário de definir um período.

print(' 1 - Rodar automático (últimos 5 dias).')
print(' 2 - Definir o intervalo de tempo.')
print(' 3 - Verificar desde o início do ano de 2019.')
print('')
opcao = int(input('Escolha uma opção (1 ou 2 ou 3):'))
print('')
if opcao == 1:
    dataApresentacaoInicio = str(date.today() - timedelta(days=5))
    dataApresentacaoFim = str(date.today())
elif opcao == 3:
    dataApresentacaoInicio = '2019-01-01'
    dataApresentacaoFim = str(date.today())
else:
    dataApresentacaoInicio = input('Digite a data início de apresentação: (Formato: AAAA-MM-DD)')
    dataApresentacaoFim = input('Digite a data final de apresentação: (Formato: AAAA-MM-DD)')

url = 'https://dadosabertos.camara.leg.br/api/v2/proposicoes?'+keywords+'dataApresentacaoInicio='+dataApresentacaoInicio+'&dataApresentacaoFim='+dataApresentacaoFim+'&ordem=ASC&ordenarPor=id'
resposta = requests.get(url)



# converter a resposta da API para dataframe
df = pd.DataFrame(resposta.json()['dados'])


print(df)

```
